```python
import json
import bs4
import os
import urllib.request as req
import requests
```

```python
#儲存url
url_list = []
response = requests.get("https://www.blocktempo.com/category/cryptocurrency-market/")
#對網頁發出請求
soup = bs4.BeautifulSoup(response.text, "html.parser")
urls = soup.find_all("div", class_="jeg_thumb")
print(f"====================第1頁====================")
for url in urls:
#title_temp 暫時儲存資料的地方
    url_temp = url.select_one("a").get("href")
#幫他放入list
    url_list.append(url_temp)


for page in range(2, 20):  # 執行1~3頁
    response = requests.get("https://www.blocktempo.com/category/cryptocurrency-market/"+"page/"+str(page)+"/")
#對網頁發出請求
    soup = bs4.BeautifulSoup(response.text, "html.parser")

#先幫他見一個list
    
#尋找要的地址位置、以及幫他放入list
    urls = soup.find_all("div", class_="jeg_thumb")
    print(f"====================第{str(page)}頁====================")
    for url in urls:
    #title_temp 暫時儲存資料的地方
        url_temp = url.select_one("a").get("href")
    #幫他放入list
        url_list.append(url_temp)
print(url_list)

import pandas as pd 
df = pd.DataFrame(url_list)
print(df)
```

```python
title_list = []

response = requests.get("https://www.blocktempo.com/category/cryptocurrency-market/")
#對網頁發出請求
soup = bs4.BeautifulSoup(response.text, "html.parser")
#先幫他見一個list
    
#尋找要的地址位置、以及幫他放入list
titles = soup.find_all("h3", class_="jeg_post_title")
for title in titles:
    #title_temp 暫時儲存資料的地方
    title_temp = title.a.text
    #######給我把那個可惡的/刪掉！！！！
    string = ''
    for char in title_temp:
        if 47 != ord(char):#用ASCII找 https://www.ascii-code.com/
            string += char 
    #幫他放入list
    title_list.append(string)
print(title_list)
print(f"====================第1頁====================")


for page in range(2, 20):  # 執行1~3頁
    response = requests.get("https://www.blocktempo.com/category/cryptocurrency-market/"+ "page/"+str(page)+"/")
#對網頁發出請求
    soup = bs4.BeautifulSoup(response.text, "html.parser")
#先幫他見一個list
    
#尋找要的地址位置、以及幫他放入list
    titles = soup.find_all("h3", class_="jeg_post_title")
    for title in titles:
    #title_temp 暫時儲存資料的地方
        title_temp = title.a.text
        
        #######給我把那個可惡的/刪掉！！！！
        string = ''
        for char in title_temp:
            if 47 != ord(char):#用ASCII找 https://www.ascii-code.com/
                string += char 
        #幫他放入list
        title_list.append(string)
    print(title_list)
    print(f"====================第{str(page)}頁====================")

import pandas as pd 
df = pd.DataFrame(title_list)
print(df)
```

```python
#可以拿來建立資料夾 但是要小心名稱可能會有問題

def mkdir(path):
    # 引入模組
    import os
 
    # 去除首位空格
    path=path.strip()
    # 去除尾部 \ 符號
    path=path.rstrip("\\")
 
    # 判斷路徑是否存在
    # 存在     True
    # 不存在   False
    isExists=os.path.exists(path)
 
    # 判斷結果
    if not isExists:
        # 如果不存在則建立目錄
        # 建立目錄操作函式
        os.makedirs(path) 
 
        print (path+'建立成功')
        return True
    else:
        # 如果目錄存在則不建立，並提示目錄已存在
        print (path+' 目錄已存在')
        return False
 
# 定義要建立的目錄
################因為這裡我不知道怎麼放入名稱
mkpath="20230312allarticle\\"
# 呼叫函式
mkdir(mkpath)
```
