```python
import json
import bs4
import os
import urllib.request as req

import requests
response = requests.get("https://www.blocktempo.com/category/cryptocurrency-market/")
#對網頁發出請求
soup = bs4.BeautifulSoup(response.text, "html.parser")
#prettify()>>將soup這個物件美化
print(soup.prettify())
```

```python
#先幫他見一個list
title_list = []
#尋找要的地址位置、以及幫他放入list
titles = soup.find_all("div", class_="jeg_thumb")
for title in titles:
    #title_temp 暫時儲存資料的地方
    title_temp = title.select_one("a").get("href")
    #幫他放入list
    title_list.append(title_temp)
print(title_list)
#alltitle = soup.find_all("h1",class_="entry-title") 
# 尋找所有在 div 標籤內 class="zIH PC4 XiG" 
#print(alltitle)
```

```python
import json
import bs4
import os
import urllib.request as req
import requests
title_list = []
for page in range(1, 6):  # 執行1~3頁
    response = requests.get("https://www.blocktempo.com/category/cryptocurrency-market/"+ str(page)+"/")
#對網頁發出請求
    soup = bs4.BeautifulSoup(response.text, "html.parser")

#先幫他見一個list
    
#尋找要的地址位置、以及幫他放入list
    titles = soup.find_all("div", class_="jeg_thumb")
    print(f"====================第{str(page)}頁====================")
    for title in titles:
    #title_temp 暫時儲存資料的地方
        title_temp = title.select_one("a").get("href")
    #幫他放入list
        title_list.append(title_temp)
    print(title_list)

import pandas as pd 
df = pd.DataFrame(title_list)
print(df)
```
