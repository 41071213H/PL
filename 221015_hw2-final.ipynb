3. 承上題，這些地區的哪一個地區最需要避難屋？ 這些地區的死亡率（人數/龍捲風個數）最高的地區（此探討之死亡率為‘粗略’計算方式）

```python
#將all裡面的東西讀取出來
temp = Data.get('all')
temp
```

```python
#基礎應用--尋找的資料。
#定義函數 find_you_want 
#利用input輸入 得到資料
things = str()
def find_you_want():
    ask = input("輸入state")
    things = input("輸入年份以及想搜尋的東西")

    for i in temp:#這邊i的型態為dict
        if ask == i.get('State'):#電腦很笨 一定要注意大小寫。
            print(ask, '龍捲風個數', i.get(things), '個')  
find_you_want()
```

```python
#定義函數＿排序 all_enumerate 排出top5
all_state_different_year_tor = []
all_state_different_year_fat = []

top5_2021_tor = []
top5_2020_tor = []
top5_2019_tor = []
top5_2018_tor = []
top5_2017_tor = []

def all_enumerate(toplist,list,state_list):
    list_enumerate = enumerate(list, start = 1)
    list_enumerate_sorted = sorted(list_enumerate, key = lambda s: s[1], reverse = True)
    k = 0
    for i in list_enumerate_sorted:
        j = i[0] - 1
        k += 1
        if k >5:
            break
        print(list[j],'in',state_list[j])
        toplist.append(state_list[j])
    print(toplist)
```

```python
#取出各地區當年的個數
#定義函數 get_each_in_list 
all_2021_tor = []
all_2020_tor = []
all_2019_tor = []
all_2018_tor = []
all_2017_tor = []

def get_each_in_list(year,things,list_a):
    for i in temp:
        list_a.append(i.get(year+things))
    print(list_a)
```

```python
# Q1
#計算全美國五年分別的(龍捲風數/死亡人數)個數
#定義函數
tornadoes = []#各年數值儲存區
fatalities = []#各年數值儲存區
str_the_five_year = []#年份順序
def sum_the_year(things,list):
    year = 2017
    summ = int()
    j = 0
    while j<5 :
        summ = 0
        for i in temp :
            years = str(year)
            summ += i.get(years+things)
        print(summ,things,'in',year)
        j += 1   
        list.append(summ)
        str_the_five_year.append(year)
        year += 1
    print(tornadoes)
```

```python
# Q1
#全美國五年分別的「龍捲風數」及 排序
sum_the_year('Tornadoes',tornadoes)
all_enumerate(all_state_different_year_tor,tornadoes,str_the_five_year)
```

```python
# Q1
#全美國五年分別的「死亡人數」及 排序
sum_the_year('Fatalities',fatalities)
all_enumerate(all_state_different_year_fat,fatalities,str_the_five_year)
```

```python
#將所有「地區」放入一個list
state = []
for i in temp:
    state.append(i.get('State'))
print(state)
```

```python
#其他練習1--計算該地區五年的總個數
#定義函數
list_tornadoes = []#各州五年加總的龍捲風數量
list_fatalities = []#各州五年加總的死亡數量
def sum_the_state(things,list):
    summ = int()
    num = 0
    for i in temp :
        
        #五年相加
        year = 2016
        summ = 0
        j=0
        while j<5 :
            year += 1
            years = str(year)
            summ += i.get(years+things)
            j += 1
        print(summ,things,'5 years','in',state[num])
        num += 1
        list.append(summ)
sum_the_state('Tornadoes',list_tornadoes)
```

```python
#其他練習1--排序該地區五年的top10
#五年最高的top10地區
all_top10 = []
list_tornadoes_enumerate = enumerate(list_tornadoes, start = 1)
list_tornadoes_enumerate_sorted = sorted(list_tornadoes_enumerate, key = lambda s: s[1], reverse = True)
k = 0
for i in list_tornadoes_enumerate_sorted:
    j = i[0] - 1
    k += 1
    if k >10:
        break
    print(list_tornadoes[j],'in',state[j])
    all_top10.append(state[j])
print(all_top10)
```

```python
# Q2
#取出各地區當年的個數
get_each_in_list("2021",'Tornadoes',all_2021_tor)
get_each_in_list("2020",'Tornadoes',all_2020_tor)
get_each_in_list("2019",'Tornadoes',all_2019_tor)
get_each_in_list("2018",'Tornadoes',all_2018_tor)
get_each_in_list("2017",'Tornadoes',all_2017_tor)
```

```python
# Q2
#排序各地區當年的個數
all_enumerate(top5_2021_tor,all_2021_tor,state)
all_enumerate(top5_2020_tor,all_2020_tor,state)
all_enumerate(top5_2019_tor,all_2019_tor,state)
all_enumerate(top5_2018_tor,all_2018_tor,state)
all_enumerate(top5_2017_tor,all_2017_tor,state)
```

```python
# Q2
#前五名交集，找出高風險區
dict_show = set(top5_2021_tor)&set(top5_2020_tor)&set(top5_2019_tor)&set(top5_2018_tor)&set(top5_2017_tor)
list_show = list(dict_show)#從dict變成list
```

```python
# Q3
#計算高風險區死亡率：找出五年間最大風險區
#定義函數 sum_which_state 
#從聯集資料中找到資料的「個數」及「人數」計算死亡率
the_rate = [] #「死亡率」輸出的list
def sum_which_state(outlist,inlist):
    for k in range(len(inlist)):
        state = inlist[k]
        sumt = int()
        sumf = int()
        year = 2016
        for i in temp:#這邊i的型態為dict
            if state == i.get('State'):#電腦很笨 一定要注意大小寫。
                #五年相加
                year = 2016
                j = 0
                while j < 5 :
                    year += 1
                    years = str(year)
                    sumt += i.get(years+'Tornadoes') #該地區五年「龍捲風個數」加總
                    sumf += i.get(years+'Fatalities') #該地區五年「死亡人數」加總
                    j += 1
                print(state,'死亡率',sumf/sumt)
                outlist.append(sumf/sumt)
sum_which_state(the_rate, list_show)
```

```python
# Q3
#排序高風險區的死亡率排名
final_sort = []
all_enumerate(final_sort,list_show,the_rate)
```
